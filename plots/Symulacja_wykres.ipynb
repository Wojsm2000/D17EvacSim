```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
df=pd.read_csv('evac_sim_data.csv')
df['iteration'] = df['time'] / 0.55
```

```python
# 1) Wczytaj obraz (screenshot)
img = mpimg.imread('evacuation_real_background.jpeg')

# 2) Załóżmy, że dokładnie na oryginale:
#    - oś X biegnie od 0 do 900 iteracji,
#    - oś Y od 0 do 220 osób.
#    dostosowujemy “extent” tak żeby osi obrazu w tle zaczynały się w punkcie (0,0) wykresu
#    Poniżej przekażemy (xmin, xmax, ymin, ymax):
extent = [-98, 900, -13, 220]

plt.figure(figsize=(10, 6))
plt.imshow(img, extent=extent, aspect='auto', origin='upper')

plt.step(
    df['iteration'],
    df['number'],
    where='post',
    color='blue',
    linewidth=2,
    label='Model'
)

# 4) Dodatkowe opisy osi i legenda
plt.xlabel('Iteration')    # lub 'Iteration (0.55 s steps)'
plt.ylabel('Number of people')
plt.title('Porównanie rzeczywistych danych (tło) z modelem')
plt.legend(loc='upper left')

plt.xlim(0, 900)     # ograniczamy oś X od 0 do 900 (jak oryginał)
plt.ylim(0, 220)     # ograniczamy oś Y od 0 do 220 (jak oryginał)

plt.tight_layout()
plt.show()
```
